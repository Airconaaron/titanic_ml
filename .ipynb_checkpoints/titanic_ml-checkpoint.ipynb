# Clearning Data

Since most ML is bassically applied linear algebra we need to transform everything into some sort of usable number.

* Turn the sexes into 1, 0
* set all NA values to the median of the set
* Those without an embark point we set to England
* turn all embark points into numbers


```python
import pandas as pd
import matplotlib.pyplot as plt

train = pd.read_csv("train.csv")
def clean_up(data):
    data["Fare"] = data["Fare"].fillna(data["Fare"].dropna().median())
    train["Age"] = data["Age"].fillna(data["Age"].dropna().median())
    
    data.loc[data["Sex"] == "male", "Sex"] = 0
    data.loc[data["Sex"] == "female", "Sex"] = 1
    
    data["Embarked"] = data["Embarked"].fillna("S")
    data.loc[data["Embarked"] == "S", "Embarked"] = 0
    data.loc[data["Embarked"] == "C", "Embarked"] = 1
    data.loc[data["Embarked"] == "Q", "Embarked"] = 2
```

```python
clean_up(train)
train[:10]
```

```python
train.Fare.plot(kind="kde")
plt.show()

train.Fare[train.Survived == 1].plot(kind="kde")
train.Fare[train.Survived == 0].plot(kind="kde")
plt.legend(["alive", "dead"])
plt.show()
```

# Decision trees

Located in scikit-learn tree module

```python
from sklearn import tree, model_selection

target = train.Survived.values

feature_list = ["Pclass", "Age", "Sex", "Fare", "Embarked", "SibSp", "Parch"]
features = train[feature_list]

decision_tree = tree.DecisionTreeClassifier(random_state = 1)

decision_tree_ = decision_tree.fit(features, target)

print(decision_tree.score(features, target))
```

```python
# Lets do a cross validation test
scores = model_selection.cross_val_score(decision_tree, features, target, scoring="accuracy", cv=50)
print(scores)
print(scores.mean())

# No better than our other model actually
```

```python
# let's view this model
tree.export_graphviz(decision_tree_, feature_names=feature_list, out_file="./decision_tree_one.dot")
```

```python
import os

os.system('dot -Tpng decision_tree_one.dot -o tree1.png')
```

![decision_tree_1](./tree1.png)

The problem with this model is that it Overfits!
This tree is trying to do too much at once.

At the very top though we can see it first splits our data by gender, then by whether someone was a child or not. Similar to our intuitions from the movie.

```python
# Fixing the tree

target = train.Survived.values

feature_list = ["Pclass", "Age", "Sex", "Fare", "Embarked", "SibSp", "Parch"]
features = train[feature_list]

decision_tree = tree.DecisionTreeClassifier(random_state = 1, 
                                               max_depth = 7,
                                                min_samples_split = 2,)

decision_tree_ = decision_tree.fit(features, target)

print(decision_tree.score(features, target))

scores = model_selection.cross_val_score(decision_tree, features, target, scoring="accuracy", cv=50)
print(scores)
print(scores.mean())

tree.export_graphviz(decision_tree_, feature_names=feature_list, out_file="./decision_tree_two.dot")

os.system('dot -Tpng decision_tree_two.dot -o tree2.png')
```

![tree_2](./tree2.png)

This tree is now much less dense.

# Extensions

Extensions to the decision tree include random forests and gradient boosted trees.
A normal decision tree simply uses the variance of each parameter as a deciding factor. It uses the column with the highest variance to first split the data, so on and so forth.

# SVM

Support vector machines in essence transform our data into a different vector space, then draws a linear boundary there. This then translates to a non-linear boundary on the original space. 

![SVM](./svm.png)

```python
from sklearn import svm

target = train.Survived.values

feature_list = ["Pclass", "Age", "Sex", "Fare", "Embarked", "SibSp", "Parch"]
features = train[feature_list]

classifier = svm.SVC(kernel='rbf') # 'poly', 'linear', 'rbf'

classifier_ = classifier.fit(features, target)
print(classifier_.score(features, target))

# Cross validation 

scores = model_selection.cross_val_score(classifier_, features, target, scoring="accuracy", cv=50)
print(scores)
print(scores.mean())

# Surprisingly terrible for an RBF kernel
```

# Neural Nets


So what happens when we put this model through into tensorflow??

![neural_nets](./neural-nets.png)

```python
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler

train = pd.read_csv("./train.csv")
clean_up(train)
scaler = MinMaxScaler()
train["Age"] = scaler.fit_transform(train["Age"].values.reshape(-1, 1))
train["Fare"] = scaler.fit_transform(train["Fare"].values.reshape(-1, 1))

print(train[:10])
feature_list = ["Pclass", "Age", "Sex", "Fare", "Embarked", "SibSp", "Parch"]
print(feature_list)


data = train.as_matrix(feature_list)

data.shape
```

```python
train_df = pd.read_csv("./train.csv")
labels = train_df.Survived.values.astype(float)
```

```python
# data split

train = data[:-100]
test = data[-100:]

train_labels = labels[:-100].reshape((-1, 1))
test_labels = labels[-100:].reshape((-1, 1))

print(train_labels[:10])
```

```python
# initialize model
tf.reset_default_graph()

features_count = 7
labels_count = 1

hidden_layer_size = 5

features_ = tf.placeholder(tf.float32, name="input")
labels_ = tf.placeholder(tf.float32, name="labels")

# weights1 = tf.Variable(tf.truncated_normal((features_count, hidden_layer_size), stddev=0.01), name="weight1")
# bias1 = tf.Variable(tf.zeros(hidden_layer_size), name="bias1")

# z1 = tf.matmul(features_, weights1) + bias1
# layer1 = tf.nn.relu(z1, name="layer1")

# weights2 = tf.Variable(tf.truncated_normal((hidden_layer_size, labels_count), stddev=0.01), name="weight2")
# bias2 = tf.Variable(tf.zeros(labels_count), name="bias2")

# logits = tf.matmul(layer1, weights2) + bias2

weights = tf.Variable(tf.truncated_normal((features_count, labels_count), stddev=0.1))
bias = tf.Variable(tf.zeros(labels_count))
logits = tf.matmul(features_, weights) + bias

output = tf.nn.sigmoid(logits)
loss = tf.losses.mean_squared_error(labels=labels_, predictions = output)

train_step = tf.train.AdamOptimizer(learning_rate = 0.01).minimize(loss)

correct_pred = tf.equal(tf.round(output), labels_)
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

```python
# training step

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(5000):
        _, loss_val, output_val = sess.run([train_step, loss, logits], feed_dict = {features_: train, labels_: train_labels})
        if i % 25 == 0:
            print(loss_val)
            print(output_val[:10])
        
        if i % 100 == 0:
            test_val = sess.run(loss, feed_dict = {features_: test, labels_: test_labels})
            print("test", test_val)
```
